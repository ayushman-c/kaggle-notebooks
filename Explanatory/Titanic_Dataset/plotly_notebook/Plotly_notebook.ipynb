```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


from google.colab import drive
drive.mount('/content/drive')


df = pd.read_csv("/content/drive/MyDrive/cleaned_titanic.csv")
```

```python
df.head()
```

```python
fig = px.histogram(df, x='Survived',
                   title='<b>Survival Distribution</b>',
                   color='Survived',
                   color_discrete_sequence=px.colors.qualitative.Safe,
                   template='plotly_white')
fig.update_layout(showlegend=False, title_x=0.5)
fig.show()
```

### Survival Overview
The dataset shows a clear imbalance: a significant number of passengers did not survive. This initial look sets the stage for investigating which factors (like class or age) might have influenced these outcomes.

### Class Distribution
It's striking to see how many more passengers were in 3rd class compared to 1st and 2nd. This distribution is key, as socio-economic status often played a major role in emergency evacuations.

```python
fig = px.scatter(df, x='Age', y='Fare',
                 color='Survived',
                 title='<b>Age vs. Fare by Survival Status</b>',
                 color_discrete_sequence=px.colors.qualitative.Bold,
                 template='plotly_white',
                 hover_data=['Pclass'])
fig.update_layout(title_x=0.5)
fig.show()
```

### Age and Fare Trends
There's a noticeable cluster of survivors among those who paid higher fares, regardless of age. We can also see a wide range of ages, but those in the high-fare bracket seem to have had a better chance.

```python
fig = px.histogram(df, x='Pclass', color='Survived',
                   facet_col='Sex',
                   title='<b>Survival Rate by Class and Gender</b>',
                   barmode='group',
                   color_discrete_sequence=px.colors.qualitative.Vivid,
                   template='plotly_white')
fig.update_layout(title_x=0.5)
fig.show()
```

### Gender and Class Impact
This breakdown really highlights the 'women and children first' policy. Females in 1st and 2nd class had very high survival rates, while males across all classes—especially 3rd—faced much tougher odds.

```python
import plotly.figure_factory as ff

corr = df.corr()
fig = ff.create_annotated_heatmap(corr.values,
                                  x=list(corr.columns),
                                  y=list(corr.index),
                                  annotation_text=corr.round(2).values,
                                  colorscale='RdBu_r')

fig.update_layout(title='<b>Feature Correlation Matrix</b>', title_x=0.5)
fig.show()
```

### Correlation Insights
The heatmap confirms a strong relationship between class and fare, and more importantly, identifies which features move together. It’s a great map for deciding which variables to use in a predictive model later.
