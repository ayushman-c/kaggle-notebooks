```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


df = pd.read_csv("/content/drive/MyDrive/cleaned_titanic.csv")
```

### 1. Overall Survival Rate
Let's start by looking at the general survival distribution.

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

**Insight:** A significant majority of passengers did not survive the disaster, showing the high fatality rate of the event.

### 2. Survival by Gender
Does gender play a role in survival rates?

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival Count by Sex")
plt.show()
```

**Insight:** Females had a much higher survival rate than males, likely due to 'women and children first' protocols.

### 3. Impact of Passenger Class
Checking how socio-economic status affected survival.

```python
sns.countplot(x="Pclass",hue="Survived",data=df)
plt.title("Survival Count by Pclass")
plt.show()
```

**Insight:** Passengers in 1st class were much more likely to survive compared to those in 3rd class.

### 4. Age Distribution
Who was on board? Let's look at the age groups.

```python
sns.histplot(df['Age'],bins=30,kde=True)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show();
```

**Insight:** The majority of passengers were young adults between 20 and 40 years old.

### 5. Survival vs Age
Did younger passengers have a better chance?

```python
sns.histplot(x="Age",data=df,hue="Survived", bins=30, kde=True)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
#
```

**Insight:** Children appear to have a visible survival spike, whereas the mortality rate was higher for middle-aged men.

### 6. Fare and Survival
Does the cost of the ticket correlate with survival?

```python
sns.boxplot(x="Survived",y="Fare", data=df)
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.title("Fare Distribution by Survival")
plt.show()
```

**Insight:** The median fare for survivors is notably higher than for those who didn't survive, suggesting a link between wealth and safety.

### 7. Feature Correlations
Looking for hidden relationships between numerical variables.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

**Insight:** Pclass and Fare show a strong negative correlation, confirming that 1st class paid significantly more.

### 8. Family Size (Siblings/Spouses)
How did traveling with family affect outcomes?

```python
sns.countplot(x="SibSp", hue="Survived", data=df)
plt.xlabel("Family Size")
plt.ylabel("Count")
plt.title("Survival Count by Family Size")
plt.show()
```

**Insight:** People traveling alone or with very large families seem to have lower survival rates than those with small families.

### 9. Interactive View: Age vs. Fare
A deeper look at the relationship between age, cost, and survival.

```python
fig = px.scatter(df, x="Age", y="Fare", color="Survived", title="Age vs. Fare")
fig.show()
```

**Insight:** High-fare passengers survived regardless of age, while low-fare survival was much more scattered.
