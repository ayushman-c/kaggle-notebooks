```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/linear regression dataset.csv")
df.shape
```

```python
df.info()
```

```python
sns.scatterplot(data=df,x='x',y='y')
```

```python
df.head(2)
```

```python
X = df.iloc[:,0]
print(X.head(2))
Y = df.iloc[:,-1]
print(Y.head(2))
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(np.array(X_train).reshape(-1,1),y_train)
```

```python
sns.scatterplot(data=df,x='x',y='y')
plt.plot(X_train,lr.predict(np.array(X_train).reshape(-1,1)),color='red')
```

```python
print(X_test.head())
print(y_test.head())
```

```python
lr.predict(X_test.iloc[4].reshape(-1,1))
```

```python
def error_percent(diff,org):
  error = (diff/org)*100
  return error

total_error_sum = 0 # Initialize the sum variable

for i in range(len(X_test)):
  predicted_value = lr.predict(X_test.iloc[i].reshape(-1,1))
  diff = abs(predicted_value - y_test.iloc[i])
  total_error_sum = total_error_sum + error_percent(diff,y_test.iloc[i]) # Use the new variable name

error_percentage = total_error_sum/len(X_test) # Renamed to avoid conflict
print(error_percentage)
```
