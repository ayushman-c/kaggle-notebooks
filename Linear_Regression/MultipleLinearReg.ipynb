```python
import pandas as pd
import numpy as np

from sklearn.datasets import make_regression #library to make linear Reg datasets
import plotly.express as px #to visualize the datasets
import plotly.graph_objects as go

from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
X,y = make_regression(n_samples = 100,
                      n_features= 2,
                      n_informative=2,
                      n_targets=1,
                      noise=10)
```

```python
df = pd.DataFrame({'col1':X[:,0],
                   'col2':X[:,1],
                   'target':y})
```

```python
df.head(2)
```

```python
fig = px.scatter_3d(df,x='col1',y='col2',z='target')
fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x='col1',
    y='col2',
    z='target',
    color='target'
)
fig.show()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
print(X_test[0:])
print(y_test[0:])
```

```python
y_pred = model.predict(X_test)
```

```python
print('Model Predictions (y_pred):')
print(y_pred)
```

### Model Evaluation Metrics

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (R2) Score: {r2:.2f}')
```

```python
# Extract coefficients and intercept from the trained model
coef_x = model.coef_[0]
coef_y = model.coef_[1]
intercept = model.intercept_

# Create a meshgrid for the plane based on the range of col1 and col2
x_range = np.linspace(df['col1'].min(), df['col1'].max(), 20)
y_range = np.linspace(df['col2'].min(), df['col2'].max(), 20)
X_plane, Y_plane = np.meshgrid(x_range, y_range)

# Calculate the Z values for the plane
Z_plane = coef_x * X_plane + coef_y * Y_plane + intercept

# Create the 3D scatter plot of the original data
scatter_plot = go.Scatter3d(
    x=df['col1'],
    y=df['col2'],
    z=df['target'],
    mode='markers',
    marker=dict(
        size=5,
        color=df['target'], # Color by target value
        colorbar=dict(title='Target Value'),
        colorscale='Viridis',
        opacity=0.8
    ),
    name='Actual Data'
)

# Create the surface plot for the regression plane
surface_plot = go.Surface(
    x=X_plane,
    y=Y_plane,
    z=Z_plane,
    opacity=0.6,
    colorscale='Plasma', # Different colorscale for the plane
    showscale=False,
    name='Regression Plane'
)

# Combine the scatter plot and the surface plot
fig = go.Figure(data=[scatter_plot, surface_plot])

fig.update_layout(
    title='3D Scatter Plot with Regression Plane',
    scene=dict(
        xaxis_title='col1',
        yaxis_title='col2',
        zaxis_title='target',
        aspectmode='cube'
    )
)

fig.show()
```
